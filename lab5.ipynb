# 绘制结果折线图

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

def draw_loss(model_type):
    data1 = np.loadtxt('./result/train_' + model_type + '_loss.txt')
    data2 = np.loadtxt('./result/valid_' + model_type + '_loss.txt')
    x  = range(0, 100)
    # plt.ylim(0.0, 1.0)
    y1 = []
    y2 = []
    for i in range(0, 100):
        y1.append(data1[i*20])
        y2.append(data2[i*5])
        
    plt.plot(x, y1, label='train_loss')
    plt.plot(x, y2, label='valid_loss')
    plt.xlabel('step')
    plt.ylabel('loss')
    plt.title(model_type +  ' Loss')
    plt.legend()
    plt.savefig('./result/' + model_type + '_loss.jpg')
    
def draw_accuracy(model_type):
    data1 = np.loadtxt('./result/train_' + model_type + '_accuracy.txt')
    data2 = np.loadtxt('./result/valid_' + model_type + '_accuracy.txt')
    x  = range(0, 100)
    # plt.ylim(0.0, 1.0)
    y1 = []
    y2 = []
    for i in range(0, 100):
        y1.append(data1[i*20])
        y2.append(data2[i*5])
        
    plt.plot(x, y1, label='train_accuracy')
    plt.plot(x, y2, label='valid_accuracy')
    plt.xlabel('step')
    plt.ylabel('accuracy')
    plt.title(model_type +  ' accuracy')
    plt.legend()
    plt.savefig('./result/' + model_type + '_accuracy.jpg')
```

```python
draw_loss('multimodal')
```

```python
draw_loss('picture')
```

```python
draw_loss('text')
```

```python
draw_accuracy('multimodal')
```

```python
draw_accuracy('picture')
```

```python
draw_accuracy('text')
```

```python
def loss(use_data):
    data1 = np.loadtxt('./result/' + use_data + '_multimodal_loss.txt')
    data2 = np.loadtxt('./result/' + use_data + '_picture_loss.txt')
    data3 = np.loadtxt('./result/' + use_data + '_text_loss.txt')
    x  = range(0, int(len(data1)/20))
    y1 = []
    y2 = []
    y3 = []
    for i in range(0, int(len(data1)/20)):
        y1.append(data1[i*20])
        y2.append(data2[i*20])
        y3.append(data3[i*20])
        
    plt.plot(x, y1, label= use_data + '_multimodal_loss')
    plt.plot(x, y2, label= use_data + '_picture_loss')
    plt.plot(x, y3, label= use_data + '_text_loss')
    plt.xlabel('step')
    plt.ylabel('loss')
    plt.title(use_data +  ' Loss')
    plt.legend()
    plt.savefig('./result/' + use_data + '_loss.jpg')
    
def accuracy(use_data):
    data1 = np.loadtxt('./result/' + use_data + '_multimodal_accuracy.txt')
    data2 = np.loadtxt('./result/' + use_data + '_picture_accuracy.txt')
    data3 = np.loadtxt('./result/' + use_data + '_text_accuracy.txt')
    x  = range(0, int(len(data1)/20))
    y1 = []
    y2 = []
    y3 = []
    for i in range(0, int(len(data1)/20)):
        y1.append(data1[i*20])
        y2.append(data2[i*20])
        y3.append(data3[i*20])
        
    plt.plot(x, y1, label= use_data + '_multimodal_accuracy')
    plt.plot(x, y2, label= use_data + '_picture_accuracy')
    plt.plot(x, y3, label= use_data + '_text_accuracy')
    plt.xlabel('step')
    plt.ylabel('accuracy')
    plt.title(use_data +  ' accuracy')
    plt.legend()
    plt.savefig('./result/' + use_data + '_accuracy.jpg')
```
